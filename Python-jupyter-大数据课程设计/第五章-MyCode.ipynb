3.查询男篮球运动员 年龄 身高 体重 三项的极差

```python
basketball=data.query("项目=='篮球'").query("性别=='男'")
```

```python
def range_data_group(arr):
    return arr.max()-arr.min()
```

```python
basketball.agg({'年龄（岁）':range_data_group,
                '身高(cm)':range_data_group,
                '体重(kg)':range_data_group,}
              )
```

4.加体质指数BMI=weight/身高^2

```python
# 先在原数据表中增加"体质指数列"
```

```python
basketball['体质指数']=0
basketball
```

```python
# 自定义一个 BMI指数计算函数
def bmi_fun(df):
    weight=df['体重(kg)']
    height=df['身高(cm)']/100
    return weight/(height**2)
```

```python
result=bmi_fun(basketball)
```

```python
result
```

```python
basketball['体质指数']=bmi_fun(basketball)
```
